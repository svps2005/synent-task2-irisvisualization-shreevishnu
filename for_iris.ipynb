Importing and 

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("IRIS.csv")
df.head()
```

```python
df.columns
```

Bar Chart : Distribution of Iris Species

```python
species_count=df["species"].value_counts()
species_count.plot(kind="bar")
plt.title("Number of floowers in each species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

Histogram: Distribution of Sepal Length

```python
plt.figure(figsize=(8,6))

plt.hist(df["sepal_length"], bins=10)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")

plt.show()
```

Scatter Plot : Relationship Between Sepal Length and Petal Length

```python
plt.figure(figsize=(8,6))

for species in df["species"].unique():
    subset = df[df["species"] == species]
    plt.scatter(
        subset["sepal_length"],
        subset["petal_length"],
        label=species
    )

plt.title("Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend()
plt.show()
```

The scatter plot compares sepal length and petal length across different Iris species. Iris-setosa forms a separate cluster, while Iris-versicolor and Iris-virginica have larger petal lengths. A positive relationship exists between sepal length and petal length.
